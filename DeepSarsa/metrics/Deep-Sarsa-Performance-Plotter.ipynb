```python
%cd /content/drive/MyDrive/P20/DeepSarsa/

import matplotlib.pyplot as plt
import pandas as pd
from lets_plot import *
from copy import deepcopy
from lets_plot import *
```

```python
deepsarsa_metrics = pd.read_pickle("./torch-deepsarsa-metrics_breakout.pkl")
linear_deepsarsa_metrics = pd.read_pickle("./torch-linear-deepsarsa-metrics_breakout.pkl")

deepsarsa_metrics = pd.DataFrame.from_dict(deepsarsa_metrics, orient='index')
linear_deepsarsa_metrics = pd.DataFrame.from_dict(linear_deepsarsa_metrics, orient='index')
```

```python
def plotter(metrics1, metrics2, model1, model2):
    metrics1_max_rolling = metrics1[["frame", "highest_score", "rolling_reward"]].melt(id_vars=["frame"], var_name="metrics")
    plot1_max_rolling = (
        ggplot(metrics1_max_rolling, aes(x="frame", y="value", color="metrics"))
        + geom_line()
        + labs(title=model1 + ": Highest Rolling Score & Rolling Reward", x="Frames", y="Scores")
        + scale_color_manual(labels=["Highest Rolling Score", "Rolling Reward"], values=["blue", "lightblue"])
    )
    display(plot1_max_rolling)
    ##ggsave(plot1_max_rolling, plot1_max_rolling.as_dict().get("ggtitle")['text']+".html") ## Uncomment to save plot

    metrics2_max_rolling = metrics2[["frame", "highest_score", "rolling_reward"]].melt(id_vars=["frame"], var_name="metrics")
    plot2_max_rolling = (
        ggplot(metrics2_max_rolling, aes(x="frame", y="value", color="metrics"))
        + geom_line()
        + labs(title=model2 + ": Highest Rolling Score & Rolling Reward", x="Frames", y="Scores")
        + scale_color_manual(labels=["Highest Rolling Score", "Rolling Reward"], values=["blue", "lightblue"])
    )
    display(plot2_max_rolling)
    #ggsave(plot2_max_rolling, plot2_max_rolling.as_dict().get("ggtitle")['text']+".html") ## Uncomment to save plot

    metrics1_hiscores = deepcopy(metrics1[["frame","highest_score"]])
    metrics2_hiscores = deepcopy(metrics2[["frame","highest_score"]])
    metrics1_hiscores["models"] = [model1]*len(metrics1_hiscores)
    metrics2_hiscores["models"] = [model2]*len(metrics2_hiscores)
    models_hi_scores = pd.concat([metrics1_hiscores, metrics2_hiscores])

    models_hi_scores = models_hi_scores.melt(id_vars=["frame", "models"], var_name="metrics")
    models_hi_scores = (
        ggplot(models_hi_scores, aes(x="frame", y="value", color="models"))
        + geom_line()
        + labs(title=model1 + " & " + model2 + ": Highest Rolling 100 Rewards", x="Frames", y="Highest Rolling 100 Rewards")
    )
    display(models_hi_scores)
    #ggsave(models_hi_scores, models_hi_scores.as_dict().get("ggtitle")['text']+".html") ## Uncomment to save plot
```

```python
plotter(deepsarsa_metrics, linear_deepsarsa_metrics, "Deep Sarsa", "Linear Deep Sarsa")
```
